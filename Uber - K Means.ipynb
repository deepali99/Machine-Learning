```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from sklearn.cluster import KMeans
```

```python
uber_data = pd.read_csv('/Users/deepali/Downloads/uber-raw-data-jun14.csv')
uber_data.head()
```

```python
lat = np.asarray(uber_data['Lat'])
lat
```

```python
lon = np.asarray(uber_data['Lon'])
lon
```

We create a numpy array of data points because the Scikit-Learn library can work with numpy array type data inputs without requiring any preprocessing.

```python
X = np.array(lat)
Y = np.array(lon)
X
```

```python
np.array((X,Y))

data = np.array((X,Y)).T
data
```

```python
plt.scatter(data[:,0],data[:,1])
plt.xticks(np.arange(39.7, 42, 0.1))
plt.yticks(np.arange(-75, -72.5, 0.1))
plt.show()
```

```python
import matplotlib.pyplot as plt
 
# Get current size
fig_size = plt.rcParams["figure.figsize"]
 
# Prints: [8.0, 6.0]
print ("Current size:", fig_size)
 
# Set figure width to 12 and height to 9
fig_size[0] = 50
fig_size[1] = 50
plt.rcParams["figure.figsize"] = fig_size
print(fig_size)
plt.show()
```

Lets take value of k = 10

Create an object of KMeans

```python
kmeans = KMeans(n_clusters= 10)
kmeans.fit(data)
```

```python
print(kmeans.cluster_centers_) 
```

Thus values of centroids is [ 40.74095248 -73.98549569] and [ 40.72495026 -73.81447788]

```python
kmeans.labels_
```

```python
plt.scatter(data[:,0],data[:,1], c=kmeans.labels_, cmap='rainbow') 
plt.show()
```
